### Define probability distributions and boxes
You should define probability distribution in this format:

$d = \{probability: reward\}$

Then you should define a new box:

$box_{n} = box(d_{n}, one ~spin ~cost)$

Let's define 3 boxes:

#### 1st box

```python
d1 = {100:0.3, 1000:0.01, 10000: 0.0001}
print(calc_stats(d1))
# print_d prints your distribution in more understandable format
print_d(d1)
```

```python
# define 1st box(distribution, cost)
box1 = box(d1, 70)
```

#### 2nd box

```python
d2 = {100:0.49}
print(calc_stats(d2))
print_d(d2)
```

```python
box2 = box(d2, 50)
```

#### 3rd box

```python
d3 = {100:0.4, 500:0.05, 1000: 0.01}
print(calc_stats(d3))
print_d(d3)
```

```python
box3 = box(d3, 100)
```

### What can you do?

You can spin it. Just call $.spin(number ~of ~spins)$ method.


```python
box1.spin(10)
```

```python
box2.spin(20)
```

```python
box3.spin(30)
```

You also can get some statistics about each box:

```python
box1.stats()
```

```python
box2.stats()
```

```python
box3.stats()
```

You can plot graphs with all statistic

```python
get_stats(box1,box2,box3)
```
